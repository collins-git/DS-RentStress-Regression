## Import Libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import os
```

## Consolidated dataset

```python
# Detect all sheets in folder. Path of folder:
directory = r'C:\Users\agustinf\OneDrive\Datasets Backup\Rent Stress - ABS scraped data (SA1)\Rent Stress Dataset in chunks'
 
# create list with all file names
list_of_files = []

# iterate over files in the folder using os
for filename in os.listdir(directory):
    file = os.path.join(directory, filename)
    # checking if it is a file just in case
    if os.path.isfile(file):
        list_of_files.append(file)
        
list_of_files
```

```python
# Iterate through all imported files and concatenate all

# Create a dataframe called abs_df
abs_df = pd.DataFrame()

# For loop
for i_file,file in enumerate(list_of_files):
    
    # Convert file into pandas DataFrame
    chunk_df = pd.read_csv(file)
    
    # Concatenate
    abs_df = pd.concat([abs_df,chunk_df])

print()
abs_df
```

```python
# # Download SA1 zone codes
url_excel = pd.read_excel(r'C:\Users\agustinf\OneDrive\Datasets Backup\ABS List of PostCodes\List of sa1.xlsx', sheet_name="SA1") 

# Change SA1 column name
url_excel.rename(columns={"Census_Code_2021":'SA1 Code'}, inplace=True)

# Change type from numeric to string as I have to concatenate to URL
url_excel['SA1 Code'] = url_excel['SA1 Code'].astype(str)

# New column "URL" concatenate SA1 code with URL structure in quick stats
url_excel["URL"] = "https://www.abs.gov.au/census/find-census-data/quickstats/2021/" + url_excel['SA1 Code']

url_excel.head()
```

```python
# Convert both columns SA1 Code into string to enable merging

url_excel["SA1 Code"] = url_excel["SA1 Code"].astype(str)
abs_df["SA1 Code"] = abs_df["SA1 Code"].astype(str)
```

```python
# Merging scraped and downloaded datapack data
complete_df = abs_df.merge(url_excel[['SA1 Code',"Area sqkm",'Region', 'State', 'SA4_NAME_2016']], on="SA1 Code", how= "inner")

print("\nShape of dataset: \n",complete_df.shape)
complete_df.head()
```

## Density column

```python
complete_df.columns
```

```python
# Convert to floats
complete_df["People"] = complete_df["People"].astype(float)
complete_df["Area sqkm"] = complete_df["Area sqkm"].astype(float)

# Create density feature
complete_df["Density"] = complete_df["People"] / complete_df["Area sqkm"]

# Drop Unnamed: 0 column
complete_df = complete_df.drop(['Unnamed: 0'], axis=1)

# Display structure of data frame
complete_df.head(2)
```

```python
# Subset only for NSW

complete_df = complete_df[complete_df["State"] == "New South Wales"]
```
