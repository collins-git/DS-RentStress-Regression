```python
list_of_features = list_of_locs[:] 

list_of_features_names = [None] * len(list_of_locs)

for i,list_code in enumerate(list_of_features):
    list_of_features_names[i] = df[list_code[0]].iloc[list_code[1],0]
    
list_of_features_names
```

```python
# Read pages in parts

url_excel[["SA1 Code","URL"]].shape[0]
```

```python
# A) Create a list with all column names
dict_for_lists = {}

for list_name in list_of_features_names:
    dict_for_lists[list_name] = []
    
dict_for_lists
```

```python

# B) For loop in 12 parts

total_rows = int(url_excel.shape[0])
chunk_size = int(total_rows/100)

print(total_rows)
print(chunk_size)

inicial_position = 0
row_list = [inicial_position]

# Load the data in chunks
for pos,row_num in enumerate(range(inicial_position, total_rows+1, chunk_size)):
    pos = pos+1
    row_list.append(row_num)
    
    list_of_code = []
    
    # Create a dictionary
    dict_for_lists = {}

    for list_name in list_of_features_names:
        dict_for_lists[list_name] = []
    

    for code,url in url_excel[["SA1 Code","URL"]].iloc[row_list[pos-1]:row_list[pos],].values:
        
        
        try:
            page = requests.get(url)
            soup = BeautifulSoup(page.content, 'html.parser')
            df = pd.read_html(page.content)
            list_of_code.append(code)
            #print(row_list[pos-1])
        except:
            continue
            
        for i,location in enumerate(list_of_locs):
            
            if list_of_features_names[i] in dict_for_lists:
                try:
                    dict_for_lists[list_of_features_names[i]].append(df[location[0]].iloc[location[1],location[2]])
                except:
                    dict_for_lists[list_of_features_names[i]].append(np.nan)
            else:
                pass

    dict_for_lists["SA1 Code"] = list_of_code
    path = r"C:\Users\agustinf\OneDrive\Datasets Backup\Rent Stress - ABS scraped data (SA1)\Rent Stress Dataset in chunks\Rent Stress last row {}.csv".format(row_list[pos])
    print(path)
    pd.DataFrame(dict_for_lists).to_csv(path)
    #print(dict_for_lists)
```
